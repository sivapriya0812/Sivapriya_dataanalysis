```python
import numpy as np
```

```python
#runs scored by 10 new players [100,50,91,78,89,25,34,19,9,10] wickets taken by same 10 new players [1,0,2,0,3,7,8,9,7,5]
#form clusters for batsmen and bowlers
runs=np.array([100,50,91,78,89,25,34,19,9,10])
wickets=np.array([1,0,2,0,3,7,8,9,7,5])
plt.scatter(runs,wickets,color="c",marker="p")
plt.show()
```

```python
#plot the scores of two students in 5subjects,subjects as X-axis and marks as Y-axis.
s1=np.array([91,92,90,95,97])
s2=np.array([84,83,89,85,92])
subjects=np.array(["telugu","english","maths","science","social"])
plt.plot(subjects,s1,label="student1")
plt.plot(subjects,s2,label="student2")
plt.title("subjects vs students")
plt.legend()
plt.show()
```

```python
stu1=[56,78,92,95,78]
stu2=[77,89,91,23,45]
sub=["Maths","Science","English","Telugu","Social"]
plt.subplot(2,1,1)
plt.plot(sub,stu1,label="Student1",color="m")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,stu2,label="Student2",color="c")
plt.legend()
```

```python
#bar graph
A=[230,560,780,127,128]
B=[200,160,270,127,400]
print("Revenues of company A",A)
print("Revenues of company B",B)
x=np.diff(A)
print("Profits of Company A",x)
y=np.diff(B)
print("Profits of Company B",y)
year=["2019-2020","2020-2021","2021-2022","2022-2023"]
plt.subplot(2,1,1)
plt.bar(year,x,label="CompanyA",color="k")
plt.legend(loc="best")
plt.subplot(2,1,2)
plt.bar(year,y,label="CompanyB",color="green")
plt.legend(loc="best")
plt.show()
```

```python
#piechart
a=np.array([25,60,5,10])
labe=["AIML","Python","Pandas","Numpy"]
plt.pie(a,labels=labe)
plt.show()
```

```python
a=np.array([25,60,5,10])
labe=["AIML","Python","Pandas","Numpy"]
explo=[0.1,0,0,0]
plt.pie(a,labels=labe,explode=explo,startangle=90,shadow="True")
plt.legend()
plt.show()
```
